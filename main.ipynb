## Train_Test_Split

```python
# adding 1's column for bias
X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
X
```

```python
Xtrain = X[0:300]
Xtest = X[300:]
ytrain = y[0: 300]
ytest = y[300:]
Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape
```

```python
Xtrain.shape
```

```python
Xtrain.T.shape
```

## Making the ML algorithm using matrix multiplication

using the formula:
$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$


```python
def HandMadeLinearRegression(X, y):
    XTX = np.dot(X.T, X)
    XTXinv = np.linalg.inv(XTX)
    model = np.dot(np.dot(XTXinv, X.T), y)
    return model

model = HandMadeLinearRegression(Xtrain, ytrain)
model
```

```python


fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(xs=df.loc[0:300, 'LSTAT'], ys=df.loc[0:300, 'RM'], zs=df.loc[0:300, 'MEDV'], c='green', label='Data')

# Create a plane
x_line = np.linspace(df.loc[0:300, 'LSTAT'].min(), df.loc[0:300, 'LSTAT'].max(), 50)
y_line = np.linspace(df.loc[0:300, 'RM'].min(), df.loc[0:300, 'RM'].max(), 50)

xx, yy = np.meshgrid(x_line,
                     y_line)

zz = model[0] + model[1]*xx + model[2]*yy   

# Plot the plane

ax.plot_surface(xx, yy, zz, color='red', alpha=0.5, label='Regression Plane')


# Labels
ax.set_xlabel('LSTAT')
ax.set_ylabel('RM')
ax.set_zlabel('MEDV')
ax.legend()

plt.show()
```

```python
# generating the predicted data
ypred = model[0] + np.dot(float(model[1]), Xtest[:, 1].reshape(206, 1)) + np.dot(float(model[2]), Xtest[:, 2].reshape(206, 1))
```

```python
# finding the MSE
(((ytest - ypred)[:, 0])**2).mean()
```
