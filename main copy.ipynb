## Train the model

```python
df
```

```python
X = df[['LSTAT', 'RM']]
X
```

```python
y = df[['MEDV']]
y
```

```python
y = y.to_numpy()
y.shape
```

```python
X = X.to_numpy()
X
```

```python
X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
X


```

```python
X[:, 0]
```

```python
X.shape
```

```python
X.T.shape
```

```python
XTX = np.dot(X.T, X)
XTX
```

```python
XTXinv = np.linalg.inv(XTX)
XTXinv
```

```python
model = np.dot(np.dot(XTXinv, X.T), y)
model
```

```python
model[0]
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(xs=df['LSTAT'], ys=df['RM'], zs=df['MEDV'], c='green', label='Data')

# Create a line
x_line = np.linspace(df['LSTAT'].min(), df['LSTAT'].max(), 50)
y_line = np.linspace(df['RM'].min(), df['RM'].max(), 50)

xx, yy = np.meshgrid(x_line,
                     y_line)

# Example: define a line function (you can replace this with a regression result)
zz = model[0] + model[1]*xx + model[2]*yy   # Example linear relation

# Plot the line

ax.plot_surface(xx, yy, zz, color='red', alpha=0.5, label='Regression Plane')


# Labels
ax.set_xlabel('LSTAT')
ax.set_ylabel('RM')
ax.set_zlabel('MEDV')
ax.legend()

plt.show()
```
